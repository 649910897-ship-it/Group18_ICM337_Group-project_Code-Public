```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
df = pd.read_excel('sweden_regression_lag2_results.xlsx')
df
```

```python
df.columns
```

```python
X = df.drop(columns=["sweden_index_return", "date"])
y = df["sweden_index_return"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    shuffle=False
)
```

```python
print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
r2_train = model.score(X_train, y_train)
r2_test = model.score(X_test, y_test)

print("Train R²:", r2_train)
print("Test R²:", r2_test)
```

```python
import statsmodels.api as sm

X_train_sm = sm.add_constant(X_train)
X_test_sm = sm.add_constant(X_test)

ols_model = sm.OLS(y_train, X_train_sm).fit()

print(ols_model.summary())
```

```python
df = pd.read_excel('COVID.xlsx')

df["COVID_period"] = (
    (df["date"] >= "2020-03-01") &
    (df["date"] <= "2021-12-31")
).astype(int)

df["log_cases"] = np.log1p(df["New Confirmed Cases"])
df["log_deaths"] = np.log1p(df["New Confirmed Deaths"])

df
```

```python
df.columns
```

```python
X = df[
    [
        "retail_sales_change_lag2",
        "repo_rate_change_lag2",
        "unemployment_change_lag2",
        "pmi_change_lag2",
        "spx_return",
        "COVID_period"
    ]
]

y = df["sweden_index_return"]

X = sm.add_constant(X)
model = sm.OLS(y, X).fit(cov_type="HC3")

print(model.summary())
```
